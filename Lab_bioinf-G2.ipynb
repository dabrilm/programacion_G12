```python
with open("/content/drive/MyDrive/Colab Notebooks/sequence_Covid-19.fasta") as f:
    lines = f.readlines()
print(lines)
```

```python
secuence=''
for l in lines[1:]:#slice, para comenzar desde el segundo elemento
  secuence+=l[:-1]
print(secuence)
```

```python
problema='TTGAAAGAGCCACCACATTTT'
a=secuence.find(problema)
print(a)
```

```python
def busqueda(lista,valor): # se usó una funcion llamada busqueda, esta se alimenta del str secuence y el str problema. En ese mismo orden

    inicio=0 # Establecimos un inicio y un final para ir viendo los caracteres por grupos de caracteres 
    final=21 # este es el final del grupo y es el largo como tal de la cadena buscada
    c=0 # esta es una variable para ir recorriendo el total del str

    while c<len(lista): # establecemos un ciclo while debido al largo indefinido de la cadena, adicionalmente por que se intentó con un for y no permitió su recorrido, quizá por mala practica
        d=lista[inicio:final] # asignamos a una variable los caracteres del grupo indicado por inicio y final 
        if d==valor: # establecemos una condicion donde compara los caracteres de la variable d con  los de valor 
            print('El comienzo de la cadena es',c,'y el final es',final-1) # por hacer mas ameno el codigo colocamos el resultado con un mensaje y los resultados 
            tupla1=(c,) # asignamos el valor de c que es el indice en el que comienza nuestra cadena buscada a la variable tupla1
            tupla2=(final-1,) # asignamos el valor del indice donde termina la cadena a la variable tupla2
            print(tupla1+tupla2) # luego sumamos las dos tuplas para obtener una tupla con ambos caracteres, el de inicio y el de final
            break # luego de esto rompemos el ciclo y salimos del ciclo y la funcion
        inicio+=1 # si los caracteres no coinciden en el if se llega a este punto y se le suma 1 a inicio y final de tal forma que se corre un caracter la secuencia
        final+=1
        c+=1 # para hacer el recorrido creamos c y la incrementamos cada vez que no encuentre el str

busqueda(secuence,problema) # luego de todo se llama a busqueda y se imprimen los resultados
```
